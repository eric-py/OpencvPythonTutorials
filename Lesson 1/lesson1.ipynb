# پردازش تصویر با استفاده از OpenCV و Matplotlib

این پروژه مثالی برای بارگذاری، نمایش و ذخیره تصاویر با استفاده از OpenCV و matplotlib در زبان پایتون ارائه می‌دهد.

## پیش‌نیازها

برای اجرای این کد، به نصب کتابخانه‌های زیر نیاز خواهید داشت:
- OpenCV: برای بارگذاری، نمایش و ذخیره تصاویر
- Matplotlib: برای نمایش تصاویر و رسم خطوط روی آنها

برای نصب این کتابخانه‌ها:
```bash
pip install opencv-python
```

````python
import cv2 as cv
````

### 1. بارگذاری تصویر
در این بخش، تصویر به فرمت خاکستری از مسیر مشخص‌شده بارگذاری می‌شود:

````python
# IMREAD_GRAYSCALE or 0: gray scale
# IMREAD_COLOR or 1: BGR(inverse RGB)
# IMREAD_UNCHANGED or -1: BGRA(inverse RGBA)

img = cv.imread('../src/peopel.png', cv.IMREAD_GRAYSCALE)
````

### 2. نمایش تصویر با OpenCV
برای نمایش تصویر در یک پنجره با OpenCV از `cv.imshow()` استفاده شده و منتظر می‌ماند تا کاربر کلیدی را فشار دهد:

````python
cv.imshow('IMAGE', img)
cv.waitKey(0) # wait for the click
cv.destroyAllWindows() # close all windows after clicking
````

### 3. ذخیره تصویر پردازش شده
تصویر خاکستری با استفاده از `cv.imwrite()` در یک فایل جدید ذخیره می‌شود:

````python
cv.imwrite('../src/final_people.png', img)
````

برای نصب matplotlib:
```bash
pip install matplotlib
```

````python
from matplotlib import pyplot as plt
````

### 4. نمایش تصویر با Matplotlib
در این بخش، تصویر با استفاده از کتابخانه matplotlib نمایش داده می‌شود. در این قسمت محورها حذف شده‌اند تا تمرکز بر روی خود تصویر باشد:

````python
plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
````

### 5. رسم خط روی تصویر
در این قسمت خطی قرمز با عرض 5 پیکسل به عنوان مثال روی تصویر کشیده شده است:

````python
plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.plot([100,100], [100,200], 'r', linewidth=5)
plt.show()
````
